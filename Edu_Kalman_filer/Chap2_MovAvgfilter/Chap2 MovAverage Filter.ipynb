```python
import Kalman_filter as fl
import numpy as np
import matplotlib.pyplot as plt

dt = 0.2
time = np.arange(0,100,dt)
i=0
Nsample = len(time)

Sample = np.zeros(Nsample)

while i < Nsample :
    Sample[i] = fl.get_static_sample(14.4,0.5)
    i=i+1

filtered_result=fl.Avgfilter(Sample) #평균필터적용
```

```python
plt.plot(time,filtered_result,'o')
plt.plot(time,Sample,'--')
plt.title("Test Average filter")
plt.xlabel("time [ms]")
plt.ylabel("Voltage [v]")

plt.show()
```
