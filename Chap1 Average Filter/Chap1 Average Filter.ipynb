# Average Filter

```python
import Avgfilter as fl
import numpy as np
import matplotlib.pyplot as plt
i=0
time=np.arange(0,100,1) # 아래의 데이터와 크기가 맞는 시간 행렬
demension = time.shape[0]
data=np.zeros(demension) 

# 특정 데이터 생성
while i < data.shape[0] :
    data[i] = 14.4 + np.random.normal(0, 0.1) # 난수 발생으로 노이즈 추가
    i=i+1
    
    
filtered_result=fl.Avgfilter(data) #평균필터적용
```

```python
plt.plot(time,filtered_result)
plt.plot(time,data)

plt.show()
```
